# Galaxy Interaction

What happens when two galaxies get too close to each other and interact?

$$ \boldsymbol{\ddot{r}}=-\gamma \left(\frac{M}{r^3}\boldsymbol{r}-\frac{S}{\rho^3}\boldsymbol{\rho}+\frac{S}{R^3}\boldsymbol{R}\right) $$

$$ \boldsymbol{\ddot{R}}=-\gamma\frac{M+S}{R^3}\boldsymbol{R} $$
